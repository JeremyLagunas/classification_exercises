# 3. Cake Recipes
### a. Read the cake_recipes table. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

```python
cake = pd.read_csv('cake_recipes.csv')
cake
```

### b. Tidy the data as necessary.

```python
cake['recipe:position'].str.split(':', expand = True)
```

```python
cake[['recipe_letter', 'position']] = cake['recipe:position'].str.split(':', expand = True)
cake
```

```python
cake_melt = cake.melt(id_vars = ['recipe_letter', 'position'], 
                     var_name = 'temp',
                     value_name = 'score')
cake_melt = cake_melt[8:]
cake_melt = cake_melt.reset_index()
cake_melt
```

### c. Which recipe, on average, is the best? recipe b

```python
cake_melt.groupby('recipe_letter').score.mean()
```

### d. Which oven temperature, on average, produces the best results? 275

```python
cake_melt.groupby('temp').score.mean()
```

### e. Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
cake_melt.groupby(['recipe_letter', 'position', 'temp']).score.mean().sort_values(ascending=False)
```
